* <strong>Primer étape <code>Estimation</code></strong>
    <p>→ Pour faire des statistiques, il faut un échantillon de données uniforme continue aléatoiressur $[a,b]$ . Et pour avoir des données sous Python, on peut fabriquer une liste contenant ces 1000 nombres</p>

```python
from random import random
def uniforme(a,b):
    return(a+(b-a)*random())
for i in range(0,1000) :
    donnees.append(uniforme(0,1))
print(donnees)
```

* <strong>2éme étape <code>Moyenne</code></strong>
<p>→ La variance est définie comme la moyenne des carrés des écarts à la moyenne:</p>

```python
def moyenne(tableau):
    return sum(tableau) / len(tableau)

print(moyenne(donnees))
```

```python
np.mean(donnees)
```

* <strong>3éme étape <code>Variance</code></strong>
<p>→ a variance est une mesure de la dispersion des valeurs d'un échantillon ou d'une distribution de probabilité:</p>

```python
np.var(donnees)
```
